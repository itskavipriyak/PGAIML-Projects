### Find the range of all the columns in the dataset

*Range = Max value - Min value (in the column)*

```python
#step 1: select only number data type to find the range.
Range = Iris.select_dtypes(include=['number'])
r = Range.max() - Range.min()

#Step 2: select only String data type. No logical operation can be performed on string. only display "max"
#class name and min class name. and concatenate at the end.
RangeStr = Iris.select_dtypes(include=['object'])
rStr = RangeStr.max()+" - "+RangeStr.min()

print("Range of all columns:\n ", r.append(rStr))
```

### Find out the IQR for the column - Age 

*Hint: IQR - Interquartile range = 75th percentile - 25th percentile* 

```python
#using percentile method, find the 1st quartile and 3rd quartile and find the IQR
Q1=np.percentile(df['Age'], 25, interpolation = 'midpoint')
print("25th percentile: ", Q1)

Q3=np.percentile(df['Age'], 75, interpolation = 'midpoint')
print("75th percentile: ", Q3)

IQR = Q3 - Q1
print("IQR of Age column: ", IQR)
```
